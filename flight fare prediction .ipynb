```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns

sns.set()
```

```python
train_data = pd.read_excel(r"Data_train.xlsx")
```

```python
train_data["Duration"].value_counts()
```

```python
# remove the null values from the dataset
train_data.dropna(inplace = True)
```

```python
train_data.isnull().sum()
```

```python
#using { to.datetime } to convert the ( object data type ) to ( timestamp ) 

train_data["Journey_day"] = pd.to_datetime(train_data["Date_of_Journey"], format = "%d/%m/%Y").dt.day


train_data["Journey_month"] = pd.to_datetime(train_data["Date_of_Journey"], format = "%d/%m/%Y").dt.month
```

```python
train_data.head()
```

```python
# drop (date_of_Journey) column because we create two related columns (Journey_day, Journey_month)

train_data.drop(["Date_of_Journey"], axis = 1, inplace = True)
```

```python
# Get the ( Hours ) and ( minutes ) from [ Departure time ] column which refers to the time that the plane leaves the gate on. Then drop [Departure time] column 

# Extracting Hours 
train_data["Dep_Hour"] = pd.to_datetime(train_data["Dep_Time"]).dt.hour

# Extracting Minutes
train_data["Dep_min"] = pd.to_datetime(train_data["Dep_Time"]).dt.minute


# drop [Dep_Time] column 
train_data.drop(["Dep_Time"], axis = 1, inplace = True)


```

```python
train_data.head()
```

```python
#Get the ( Hours ) and ( minutes ) from [ Arrival Time ] column which refers to the time that the plane arrived in it's destination, Then drop [Arrival Time] column 

# Arrival Hours 
train_data["Arrival_hour"] = pd.to_datetime(train_data.Arrival_Time).dt.hour

# Arrival Minutes 
train_data["Arrival_min"] = pd.to_datetime(train_data.Arrival_Time).dt.minute

# drop [ Arrival Time] column
train_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
# converting Duration column to list
duration  = list(train_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) !=2:  # check if duration column contains only hours or minutes " if duration lenght is not equal to two"
        
        # if Hours founded add ( 0 minutes ) to it and else, add ( 0 hours )
        if "h" in duration[i]:     
            duration[i] = duration[i].strip() + " 0m" 
        else : 
                duration[i] = "0h " + duration[i]
    # append (duration_hours, duration_mins) back again            
duration_hours = []
duration_mins = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep = "h")[0]))
    duration_mins.append(int(duration[i].split(sep = "m")[0].split()[-1]))
    
    
    

```

```python
# Adding duration_hours, duration_mins list to train_data

train_data["Duration_hours"] = duration_hours
train_data["Duration_mins"] = duration_mins
```

```python
train_data["Airline"].value_counts()
```

```python
# drop Duration column from the dataset

train_data.drop(["Duration"], axis = 1, inplace = True)
```

```python
# using (OneHotEncoding) because ( Airline ) column is a ( nominal categorical)

Airline = train_data[["Airline"]]

# get_dummies with OneHotEncoding and drop the first
Airline = pd.get_dummies(Airline, drop_first = True)
Airline.head()

```

```python
# plot ( Source, Price )
sns.catplot(y = "Price", x = "Source", data = train_data.sort_values("Price", ascending = False), kind="boxen", height = 6, aspect = 3)
plt.show()
```

```python
# using (OneHotEncoding) because ( Source ) column is a ( nominal categorical)

Source = train_data[["Source"]]

Source = pd.get_dummies(Source, drop_first= True)

Source.head()
```

```python
# Destination col

train_data["Destination"].value_counts()
```

```python
# using (OneHotEncoding) because ( Destination ) column is a ( nominal categorical)

Destination = train_data[["Destination"]]

Destination = pd.get_dummies(Destination, drop_first = True)

Destination.head()
```

```python
train_data["Route"]
```

```python
# drop (Route) column because it's related to ( Total_Stops ) column 
train_data.drop(["Route"], axis = 1, inplace = True)


```

```python
# drop ( Additional Info ) column because it's only have 20% of info inside it 

train_data.drop(["Additional_Info"], axis = 1, inplace = True)
```

```python
train_data["Total_Stops"].value_counts()
```

```python
# replace values in (Total_Stops) column with numbers (non-stop) = 0, (1 stop) = 1, ...
train_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
# concatenote the training data features together 
data_train = pd.concat([train_data, Airline, Source, Destination], axis = 1)
```
